# Adding datasets

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import numpy as np
import plotly.express as px
```

```python
# Importing the Stock Price Data:
stocks_df = pd.read_csv("../data/stock_prices.csv", header = None, parse_dates=[3])
stocks_df.columns = ["id", "company_id", "stock_ticker", "date", 
                     "u_close", "u_open","u_high", "u_low", "u_volumn", 
                     "close", "open", "high", "low",  "volume", "change", 
                     "change_percent","label", "change_over_time"]
```

```python
#importing tweeter dataset
twitter_df = pd.read_csv("../data/twitter_sentiment.csv", parse_dates=['date'])
```

```python
#auxiliary datasets
company_ids = twitter_df.company_id.unique()
#stocks_df = stocks_df[stocks_df.company_id.isin(company_ids)].dropna()
company_ids_stocks = stocks_df.company_id.unique()
```

```python
#keeping only the ones in common
stocks_df = stocks_df[stocks_df["company_id"].isin(company_ids)]
stocks_df.reset_index()
twitter_df = twitter_df[twitter_df["company_id"].isin(company_ids)]
```

```python
#dropping unecessary columns
twitter_df = twitter_df.drop(columns=['Unnamed: 2', "Unnamed: 3", "Unnamed: 4"])
```

```python
twitter_df = twitter_df.drop(columns=['Unnamed: 0', "index"])

twitter_df.reset_index(inplace = True, drop = True)
```

```python
#stock data just from 2017
stocks_df = stocks_df[stocks_df["date"] >= "2017-01-01"]
```

## Adding the mean scores to the stock dataset

### Standard and weigthed mean

```python
#preparing final dataset
stocks_data_twitter = stocks_df[["company_id", "date", "close", "change", "change_percent", "change_over_time"]]
```

```python
twitter_df = twitter_df.reset_index()
```

```python
twitter_df_mean = twitter_df.groupby(['company_id', "date"]).agg(
    #weigthed_mean_overall_sentiment = ("overall_sentiment", lambda x: np.average(x, weights = twitter_df.loc[x.index, "favorites"])),
    mean_overall_sentiment = ("overall_sentiment", "mean"),
    #weigthed_mean_positive_sentiment = ("positive_sentiment", lambda x: np.average(x, weights = twitter_df.loc[x.index, "favorites"])),
    mean_positive_sentiment = ("positive_sentiment", "mean"),
    #weigthed_mean_negative_sentiment = ("negative_sentiment", lambda x: np.average(x, weights = twitter_df.loc[x.index, "favorites"])),
    mean_negative_sentiment = ("negative_sentiment", "mean"),
    #weigthed_mean_neutral_sentiment = ("neutral_sentiment", lambda x: np.average(x, weights = twitter_df.loc[x.index, "favorites"])                                      ),
    mean_negative_neutral_sentiment = ("neutral_sentiment", "mean"),
    favories = ("favorites", sum),
    retweets = ("retweets", sum),
    company_name = ("company_name", "first"),
    location = ("location", "first"), 
    district = ("district", "first"),
    industry_id = ("industry_id", "first"),
    industry_name = ("industry_name", "first"),
    clean_length = ("clean_length", "mean"),
    week = ("week", max),
    month = ("month", max),
    year = ("year", max))
    
```

```python
twitter_df_mean = twitter_df_mean.reset_index()
```

```python
#adding the weighted mean 
#adding the mean for the sentiment
twitter_df["favorites"] = twitter_df["favorites"].replace(0, 1)

wm = lambda x: np.average(x, weights = twitter_df.loc[x.index, "favorites"])

f = {"overall_sentiment": {"wm": wm}, "positive_sentiment": {"wm": wm}, "negative_sentiment": {"wm": wm}, "neutral_sentiment": {"wm": wm}}

twitter_df_wigthed_mean = twitter_df.groupby(['company_id', "date"]).agg(f)
```

```python
twitter_df_wigthed_mean.reset_index()
```

## Combining means with stock prices 

```python
#combining both 
twitter_df_agg = pd.merge(twitter_df_mean, twitter_df_wigthed_mean, on=['company_id', 'date'])
```

```python
stocks_data_twitter = pd.merge(twitter_df_agg, stocks_data_twitter, on=['company_id', 'date'])
```

```python
# changing column names
stocks_data_twitter = stocks_data_twitter.rename(columns ={('overall_sentiment', 'wm') : "wm_overall_sentiment" ,('positive_sentiment', 'wm') : "wm_positive_sentiment", ('negative_sentiment', 'wm') : "wm_negative_sentiment", ('neutral_sentiment', 'wm') : "wm_neutral_sentiment"})
```

```python
stocks_data_twitter.to_csv('../data/stocks_data_twitter.csv',encoding='utf-8')
```
